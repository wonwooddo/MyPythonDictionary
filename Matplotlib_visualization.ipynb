```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```
x=np.linspace(0, 10, 100)
```

```
fig, ax = plt.subplots(2, 3)
fig.suptitle('test')
ax[0, 0].plot(np.sin(x))
ax[0, 1].plot(x, np.sin(x), '--c'); ax[0, 1].plot(x+5, np.sin(x), ':r')
ax[0, 2].plot(x, np.sin(x), ':r', )

ax[1, 0].plot(x, np.sin(x), '--c')
ax[1, 1].plot(x, np.sin(x), '--c');ax[1, 1].set_xlabel("x");ax[1, 1].set_ylabel("y")
ax[1, 2].plot(x, np.sin(x), ':r') ;ax[1, 2].set_xlim(0, 5); ax[1, 2].set_ylim(-1, 2)

fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
